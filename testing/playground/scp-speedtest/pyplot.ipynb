```python
import datetime
out_path = '/tf/aqumon/test-scp-2.out'
f = open(out_path, 'r')
f_content = f.read()

line_array = list(f_content.split('\n'))
oneM = {'start_time':[], 'upload_speed':[], 'download_speed':[]}
fiveM = {'start_time':[], 'upload_speed':[], 'download_speed':[]}
tenM = {'start_time':[], 'upload_speed':[], 'download_speed':[]}
fiftyM = {'start_time':[], 'upload_speed':[], 'download_speed':[]}
oneHM = {'start_time':[], 'upload_speed':[], 'download_speed':[]}
twoHM = {'start_time':[], 'upload_speed':[], 'download_speed':[]}
fiveHM = {'start_time':[], 'upload_speed':[], 'download_speed':[]}
tenHM = {'start_time':[], 'upload_speed':[], 'download_speed':[]}
HM_15 = {'start_time':[], 'upload_speed':[], 'download_speed':[]}

for i,x in enumerate(line_array):
    if (i%5) == 0:
        start_time = x.split(': ')[1].split(' CST')[0] # time string
        start_dt = datetime.datetime.strptime(start_time+" 2022", '%a %b %d %H:%M:%S %Y')
        upload_speed = 0 # KBps
        download_speed = 0 # KBps
        key_dict = x.split(': ')[-1]
    elif (i%5) == 2:
        upload_speed = float(int(x.split('Upload speed:')[1].split('KBps')[0])/1000) # KBps
    elif (i%5) == 3:
        download_speed = float(int(x.split('Download speed:')[1].split('KBps')[0])/1000) # KBps
    elif (i%5) == 4:
        try:
            if key_dict == '1 M':
                # print(start_dt, upload_speed, download_speed)
                oneM['start_time'].append(start_dt)
                oneM['upload_speed'].append(upload_speed)
                oneM['download_speed'].append(download_speed)
            elif key_dict == '5 M':
                # print(start_dt, upload_speed, download_speed)
                fiveM['start_time'].append(start_dt)
                fiveM['upload_speed'].append(upload_speed)
                fiveM['download_speed'].append(download_speed)
            elif key_dict == '10 M':
                # print(start_dt, upload_speed, download_speed)
                tenM['start_time'].append(start_dt)
                tenM['upload_speed'].append(upload_speed)
                tenM['download_speed'].append(download_speed)
            elif key_dict == '50 M':
                # print(start_dt, upload_speed, download_speed)
                fiftyM['start_time'].append(start_dt)
                fiftyM['upload_speed'].append(upload_speed)
                fiftyM['download_speed'].append(download_speed)
            elif key_dict == '100 M':
                # print(start_dt, upload_speed, download_speed)
                oneHM['start_time'].append(start_dt)
                oneHM['upload_speed'].append(upload_speed)
                oneHM['download_speed'].append(download_speed)
            elif key_dict == '200 M':
                # print(start_dt, upload_speed, download_speed)
                twoHM['start_time'].append(start_dt)
                twoHM['upload_speed'].append(upload_speed)
                twoHM['download_speed'].append(download_speed)
            elif key_dict == '500 M':
                # print(start_dt, upload_speed, download_speed)
                fiveHM['start_time'].append(start_dt)
                fiveHM['upload_speed'].append(upload_speed)
                fiveHM['download_speed'].append(download_speed)
            elif key_dict == '1000 M':
                # print(start_dt, upload_speed, download_speed)
                tenHM['start_time'].append(start_dt)
                tenHM['upload_speed'].append(upload_speed)
                tenHM['download_speed'].append(download_speed)
            elif key_dict == '1500 M':
                # print(start_dt, upload_speed, download_speed)
                HM_15['start_time'].append(start_dt)
                HM_15['upload_speed'].append(upload_speed)
                HM_15['download_speed'].append(download_speed)
        except:
            continue
```
